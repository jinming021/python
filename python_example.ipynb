```python
#1)引用
a = [1,2,3,[4,5,6]]
b = a
print(a)
print(b)

a[3] = 100

print(a)
print(b)
```

```python
#2)浅拷贝Copy
a = [1,2,3,[4,5,6]]
b = a.copy()
print(a)
print(b)

a[3] = 100

print(a)
print(b)
```

```python
#3)Copy模块 - 浅拷贝
import copy
a = [1,2,3,[4,5,6]]
b = copy.copy(a)
print(a)
print(b)

a[3] = 100

print(a)
print(b)
```

```python
#4)Copy模块 - 深拷贝
import copy
a = [1,2,3,[4,5,6]]
b = copy.deepcopy(a)
print(a)
print(b)

a[3] = 100

print(a)
print(b)
```
