```python

"""
0. 打印杨辉三角形
杨辉三角形是中国古代数学的杰出研究成果之一，是我国北宋数学家贾宪于 1050 年首先发现并使用的。而后南宋数学家杨辉在《详解九章算法》一书中记载并保存了“贾宪三角形“。因此，贾宪三角形又被称为杨辉三角形。
简单地说，杨辉三角形的构成如下面动图所示：

从上面的动图我们可以得出一个道公式：杨辉三角形第i行j列的值 = 第 (i-1) 行 (j-1) 列的值 + 第 (i-1) 行 (j) 列的值
好，大家知道它的构成形式就可以开干了，千万不要以“我不懂数学，没听说过杨辉三角形”为理由放弃对本题目的挑战哦^o^
好了，大家开拔吧！
"""

# 初始化杨辉三角形
# 创建一个10*10的二维列表，并将所有的元素初始化为0
tri = []

for i in range(10):
    tri.append([])
    for j in range(10):
        tri[i].append(0)
print(tri)

# 计算杨辉三角形
# 根据观察，我们知道杨辉三角形左右两边的元素均为1
for i in range(10):
    tri[i][0] = 1
    tri[i][i] = 1
print(tri)

# 第i行j列的值 = 第(i-1)行(j-1)列的值 + 第(i-1)行(j)列的值
for i in range(2,10):
    for j in range(1,i):
        tri[i][j] = tri[i-1][j-1] + tri[i-1][j]
print(tri)


# 输出杨辉三角形

for i in range(10):
    for j in range(i+1):
        print(tri[i][j],end=" ")
    print()


#1. 更改杨辉三角形的打印样式
for i in range(10):
    for k in range((10-i)//2):
        print("\t",end=" ")
    for j in range(i+1):
        if i%2==1:
            print(" ",end="")
        print(tri[i][j],end="\t")
    print()

#等待更加牛逼的我
```

```python
class Shape:
    def __init__(self,name):
        self.name = name
    def area(self):
        pass

class Square(Shape):
    def __init__(self,length):
        super().__init__("正方形")
        self.length = length
    def area(self):
        return self.length * self.length

class Circle(Shape):
    def __init__(self,radius):
        super().__init__("圆")
        self.radius = radius
    def area(self):
        return 3.14*self.radius * self.radius

class Triangle(Shape):
    def __init__(self,base,height):
        super().__init__("三角形")
        self.base = base
        self.height = height
    def area(self):
        return self.base * self.height / 2

s = Square(4)
c = Circle(5)
t = Triangle(4,5)

print(s.name)
print(c.name)
print(t.name)

print(s.area())
print(c.area())
print(t.area())


        
```
